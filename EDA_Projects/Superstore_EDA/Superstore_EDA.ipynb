```python
####Importing packages and importing dataset for exploratory data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/Users/jeeva/Downloads/Power BI/archive/Superstore_Cleaned.csv', encoding='latin1')

print("Initial Dataset Shape:",df.shape)
print(df.head())
print(df.info())
```

```python
###Checking the histogram so that data distribution could be understood among the numerical values
print(df.hist(figsize=(12,8)))
```

```python
####checking which region has the high impact on entries
df['Region'].value_counts().plot(kind='bar')
```

```python
####checking which ship mode has the high impact 
df['Ship Mode'].value_counts().plot(kind='bar')
```

```python
####Correlation analysis to undertstand the relationship between two variables
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

```python

####Filerting orders by month and figuring out the trend 
df['Order Date'] = pd.to_datetime(df['Order Date'],format='mixed',)
sales_by_month = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

sales_by_month.plot(figsize=(12,6))
plt.title("Monthly Sales Trend")
plt.show()
```

```python
####re-check of outliers after cleaning the dataset
sns.boxplot(x=df['Profit'])
```

```python
####checking which category pulls more revenue
df.groupby('Category')['Profit'].sum().plot(kind='bar')
```
